```python
import gym
import numpy as np
import matplotlib.pyplot as plt
from qlearn import Qagent
```

```python
env = gym.make('CartPole-v1')
```

```python
bounds = [2.4, 5, 15*np.pi/360, 5]
f_env = [2, 5, 25, 15]
#f_env = 50
f_act = 2
agent = Qagent(bounds, f_env, f_act)
```

```python
agent.fit(env, epochs=20000, alpha=0.05, epsilon=1, gamma=0.9, T=1.5)
```

```python
w = 30
time = 200000
ma = np.convolve(agent.lengths_games, np.ones(w), 'valid') / w
plt.plot(agent.lengths_games[:time])
plt.plot(ma[:time])
plt.grid()
plt.show()
```
